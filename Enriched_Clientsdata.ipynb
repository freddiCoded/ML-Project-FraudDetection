```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt # run pip install plotly if error
import seaborn
import seaborn as sns
import plotly.express as px # if error install in terminal : pip install plotly
import datetime as dt
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline # show the plots on the online copy
```

```python
clients = pd.read_csv('data/client_train.csv')
invoices = pd.read_csv('data/invoice_train.csv')
```

```python
clients.head()
```

```python
# Transforming to date.time type
clients['creation_date'] = pd.to_datetime(clients['creation_date'], format='%d/%m/%Y')
```

```python
# Change datatypes
clients = clients.astype({'disrict': str})
clients = clients.astype({'client_catg': str})
clients = clients.astype({'region': str})
```

```python
dist_cols = pd.get_dummies(clients['disrict'], prefix='district_', drop_first=True)
dist_cols1 = pd.get_dummies(clients['client_catg'], prefix='client_catg_', drop_first=True)
dist_cols2 = pd.get_dummies(clients['region'], prefix='region_', drop_first=True)
```

```python
clients = pd.concat([clients, dist_cols, dist_cols1, dist_cols2], axis=1)
```

```python
clients.head()
```

```python
clients = clients.drop(['disrict', 'client_catg', 'region'], axis=1)
```

```python
clients.info()
```

```python
clients.head()
```

```python
from datetime import datetime
datetime_str = "2019-12-31"
datetime_object = datetime.strptime(datetime_str, '%Y-%m-%d')
```

```python
clients["duration"] = datetime_object - clients["creation_date"]
clients["duration"] = clients["duration"].dt.days.astype('int16')
clients.duration.head()
```

```python
clients.head()
```

```python
invoices.info()
```

```python
df_agg = invoices[["client_id", "consommation_level_1", "consommation_level_2","consommation_level_3","consommation_level_4"]].groupby("client_id").agg(["mean","std","min","max"]).reset_index()
```

```python
df_agg.columns = ['_'.join(col) for col in df_agg.columns.values]
```

```python
df_agg.head()
```

```python
df_agg = df_agg.dropna()
```

```python
df_agg.info()
```

```python
df_agg.rename(columns={"client_id_": "client_id"},inplace = True)
```

```python
clients = pd.merge(clients, df_agg, how='left', on='client_id')
```

```python
# Before we go on let's split the data into features and target label
X = clients.drop(['target','creation_date'], axis=1)
y = clients['target']# Y
```

```python
# Import train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y ,test_size=0.3, random_state=150)
# Split the 'features' and 'income' data into training and testing sets

# Show the results of the split
print ("Training set has {} samples.".format(X_train.shape[0]))
print ("Testing set has {} samples.".format(X_test.shape[0]))
```
